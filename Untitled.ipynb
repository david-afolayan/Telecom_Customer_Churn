```python
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(percent_missing*100)))
```

```python
# Summary statistics for numerical columns
summary_stats = df.describe()

# Summary of the churn status
City_counts = df['City'].value_counts()

# Display the results
print(summary_stats)
print(City_counts)

#32.555828 -124.240051
```

```python
# Filter out entries with negative monthly charges
negative_monthly_charges = df[df['Monthly Charge'] < 0]

# Display these entries
print(negative_monthly_charges)
```

```python
# Calculate the churn rate
churn_rate = (df['Customer Status'] == 'Churned').mean()

# Display the churn rate
print(churn_rate)
```

```python
sns.set(style="whitegrid")

# City Distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Gender',hue='Gender', palette='viridis')
plt.title('Evaluation of the Most Patronage Gender', size= 20)
plt.xlabel('Gender', size= 15)
plt.ylabel('Count of Gender', size= 14)
plt.show()
```

```python
# Age Distribution by Churn Status
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', hue='Customer Status', multiple='stack', palette='viridis', bins=30)
plt.title('Age Distribution by Customer Status', size= 20)
plt.xlabel('Age', size= 17)
plt.ylabel('Frequency', size= 14)
plt.show()
```

```python
# Monthly Charges by Churn Status
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Customer Status', y='Monthly Charge', hue='Customer Status', palette='viridis')
plt.title('Monthly Charges by Customer Status')
plt.xlabel('Customer Status')
plt.ylabel('Monthly Charge')
plt.show()
```

```python
# Total Charges by Churn Status
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Customer Status', y='Total Charges', hue= 'Customer Status', palette='viridis')
plt.title('Total Charges by Customer Status')
plt.xlabel('Customer Status')
plt.ylabel('Total Charges')
plt.show()
```

```python
sns.regplot(x="Latitude", y="Longitude", data=df, scatter_kws = {"color": "red"},line_kws = {"color": "blue"}).set_title('Latitude vs Longitude')
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

# Sorting the correlation pairs
sorted_pairs = corr_pairs.sort_values()

# Including only pairs that have a correlation greater than 0.5.
# A correlation coefficient of .50 or larger represents a strong or large correlation.
high_corr = sorted_pairs[(sorted_pairs) > 0.5]  

high_corr
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')
plt.figure(figsize=(20, 5))
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation Matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```
